```
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```
headers={'User-Agent': 'Morzilla/5.0 (Windows NT 6.3;Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
```

```
data=requests.get('https://www.ambitionbox.com/list-of-companies?page=1',headers=headers).text
```

```
data
```

```
soup=BeautifulSoup(data,'lxml')
```

```
company=soup.find_all('div',class_='company-content-wrapper')
```

```
len(company)
```

```
name=[]
rating=[]
reviews=[]
ctype=[]
hq=[]
old=[]
employees=[]
for i in company:
  name.append(i.find('h2').text.strip())
  rating.append(i.find('p',class_='rating').text.strip())
  reviews.append(i.find('a',class_='review-count').text.strip())
  ctype.append(i.find_all('p',class_='infoEntity')[0].text.strip())
  hq.append(i.find_all('p',class_='infoEntity')[0].text.strip())
  old.append(i.find_all('p',class_='infoEntity')[0].text.strip())
  employees.append(i.find_all('p',class_='infoEntity')[0].text.strip())

d={'name':name, 'rating':rating, 'reviews':reviews, 'ctype':ctype, 'hq':hq, 'old':old, 'employees':employees} 
df=pd.DataFrame(d)
df  
```

```

df['comb']=str(df['name'])+''+str(df['rating'])+''+str(df['reviews'])+''+str(df['ctype'])+''+str(df['hq'])+''+str(df['old'])+''+str(df['employees'])
df['comb']
```

```
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```
count_vect=CountVectorizer()
```

```
count_matrix=count_vect.fit_transform(df['comb'])
count_matrix
```

```
similar=cosine_similarity(count_matrix)
```

```
similar
```

```
company_index=df.loc[df['name']=='TCS'].index[0]
```

```
le=list(enumerate(similar[company_index]))
```

```
le
```

```
lmt=sorted(le,key=lambda x:x[1],reverse=True)
```

```
len(lmt[1:9])
```

```
lmt[1:12]
```

```
top_10=lmt[1:10]
```

```

company=[]
for i in range(len(top_10)):
  a=top_10[i][0]
  company.append(df['name'][a])
```
